```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# import seaborn as sns
# sns.set()
import datetime 
from matplotlib import style
import pandas_datareader.data as web
style.use('ggplot')
```

```python
start_date = datetime.datetime(2019, 4, 16)
end_date = datetime.datetime(2021, 4, 15)
dataframe_TSLA = web.DataReader('TSLA','yahoo', start_date, end_date)
dataframe_GOOG = web.DataReader('GOOG','yahoo', start_date, end_date)
dataframe_SPY = web.DataReader('SPY','yahoo', start_date, end_date)
dataframe_GLD = web.DataReader('GLD','yahoo', start_date, end_date)
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize = (12, 12))
ax1.plot(dataframe_TSLA.index, dataframe_TSLA['Adj Close'], 'tab:orange', linewidth = 2)
ax1.set_title('TSLA', fontsize = 18)
ax2.plot(dataframe_GOOG.index, dataframe_GOOG['Adj Close'], 'tab:orange', linewidth = 2)
ax2.set_title('GOOG', fontsize = 18)
ax3.plot(dataframe_SPY.index, dataframe_SPY['Adj Close'], 'tab:orange', linewidth = 2)
ax3.set_title('SPY', fontsize = 18)
ax4.plot(dataframe_GLD.index, dataframe_GLD['Adj Close'], 'tab:orange', linewidth = 2)
ax4.set_title('GLD', fontsize = 18)
fig.tight_layout()

for ax in (ax1, ax2, ax3, ax4):
    ax.patch.set_facecolor('black')
    ax.set(xlabel = 'Dates', ylabel = 'Adj Close')
    ax.label_outer()

plt.savefig('Adj_Close_raw_fig.jpg')
```

```python
def random(x):
    return np.random.normal(scale = 100, size = x)

plt.scatter(dataframe_TSLA['Adj Close'], dataframe_GOOG['Adj Close'], c = random(504), marker='o')
plt.title('TSLA GOOG Correlation')
plt.xlabel('TSLA')
plt.ylabel('GOOG')
plt.savefig('TSLA_GOOG_Corr_fig.jpg')
```
